```python
import matplotlib.pyplot as plt
import numpy as np
from nba_py import player
%matplotlib inline
```

```python
players = [('Clint',  'Capela'), ('Andre', 'Drummond'), ('Dwight', 'Howard'), ('DeAndre', 'Jordan')]
```

```python
players_data = []
for curr_player in players:
    player_id = player.get_player(curr_player[0], curr_player[1], season="2017-18")
    players_data.append((curr_player[0] + " " + curr_player[1], player.PlayerCareer(player_id).regular_season_totals()))
```

```python
last_year, this_year, players = [], [], []
for player_data in players_data:
    data = player_data[1]
    players.append(player_data[0])
    last_year.append(data.loc[data.SEASON_ID == '2016-17'].FT_PCT.iloc[0])
    this_year.append(data.loc[data.SEASON_ID == '2017-18'].FT_PCT.iloc[0])
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(last_year, players, color="#53B263", s=200, label="FT% in 2016-17")
plt.scatter(this_year, players, color="#2467D1", s=200, label="FT% in 2017-18")
plt.hlines(players, last_year, this_year, colors=["#A1A9B1"])
plt.title("FT% change for Howard, Jordan, Capela and Drummond", fontsize=16)
plt.xlabel("Free Throw percentage", fontsize=14)
plt.tick_params(axis='both', labelsize=12)
plt.legend(fontsize=12, bbox_to_anchor=(1.22,1))
plt.savefig('ft_change.png', bbox_inches='tight')
```
