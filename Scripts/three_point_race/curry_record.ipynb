```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from nba_py import player
```

```python
players = [
    'Ray_Allen',
    'Reggie_Miller',
    'Jason_Terry',
    'Paul_Pierce',
    'Kyle_Korver',
    'Jamal_Crawford',
    'Vince_Carter',
    'Stephen_Curry'
]
colors = [
    '#4286f4',
    '#a85f00',
    '#41f4a0',
    '#007c13',
    '#7c000e',
    '#902cd3',
    '#4f6087',
    '#FFCD34'
]
```

```python
folder = '../../Data/three_point_race/career_three_points/'
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(12,10))
for i, file_name in enumerate(players):
    player_name = file_name.split("_")[0] + " " + file_name.split("_")[1]
    current_df = pd.read_csv(folder + file_name + ".csv")
    plt.plot(current_df.GAME_CAREER, current_df.TOTAL_3MADE, c=colors[i], label=player_name, linewidth=3.0)

plt.legend(loc="upper right", bbox_to_anchor=(1.25,1))
plt.title("Three pointers made for top 8 shooters through career", fontsize=18)
plt.text(x=-60, y=-400, s="instagram.com/bballytics", fontsize=13)
plt.text(x=1120, y=-400, s="Source: basketball-reference", fontsize=13)
plt.xlabel("Game in career", fontsize=16)
plt.ylabel("Total three pointers", fontsize=16)
plt.savefig('shooters_comparison.png', bbox_inches='tight')
```
