```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
harden_2017_18 = pd.read_csv('../../Data/three_point_race/harden_2017_18.csv')
steph_2015_16 = pd.read_csv('../../Data/three_point_race/steph_2015_16.csv')
steph_2016_17 = pd.read_csv('../../Data/three_point_race/steph_2016_17.csv')
```

```python
def sum_three_pointers(df):
    total_sum = 0
    sum_three_pointers = []
    for i in range(len(df)):
        total_sum += df.iloc[i]['3P']
        sum_three_pointers.append(total_sum)
    df['SUM_3P'] = sum_three_pointers
    return df
```

```python
harden_2017_18 = sum_three_pointers(harden_2017_18)
steph_2015_16 = sum_three_pointers(steph_2015_16)
steph_2016_17 = sum_three_pointers(steph_2016_17)
```

```python
plt.figure(figsize=(16, 16), facecolor='white')

plt.plot(harden_2017_18.G, harden_2017_18.SUM_3P, color="#CE1141", label="3PM by James Harden in 2017-18 (90 so far)")
plt.plot(steph_2015_16.G, steph_2015_16.SUM_3P, color="#243E90", label="3PM by Stephen Curry in 2015-16 (402 total)")
plt.plot(steph_2016_17.G, steph_2016_17.SUM_3P, color="#FFCD34", label="3PM by Stephen Curry in 2016-17 (324 total)")
plt.xticks(np.arange(0, 21, 2))
plt.yticks(np.arange(0, 111, 10))
plt.ylim(0, 110)
plt.xlim(0, 20)
plt.xlabel("Game of season", fontsize=16)
plt.ylabel("Total 3PM through game", fontsize=16)
plt.title("Comparison of 3PM through first 20 games for Harden in 17-18 and Steph's best seasons", fontsize=20)

plt.legend(loc="upper right", bbox_to_anchor=(1.43, 1), fontsize=16)

max_harden, max_steph_15_16, max_steph_16_17 = harden_2017_18['3P'].argmax(), steph_2015_16['3P'].argmax(), steph_2016_17['3P'].argmax()

plt.text(x=max_harden-1.15, y=harden_2017_18.iloc[max_harden].SUM_3P + 1, s=(str(harden_2017_18.iloc[max_harden]['3P']) + " vs " + str(harden_2017_18.iloc[max_harden]['Opp'])), color="#CE1141", fontsize=12)
plt.text(x=max_steph_15_16-1.15, y=steph_2015_16.iloc[max_steph_15_16].SUM_3P + 1, s=(str(steph_2015_16.iloc[max_steph_15_16]['3P']) + " vs " + str(steph_2015_16.iloc[max_steph_15_16]['Opp'])), color="#243E90", fontsize=12)
plt.text(x=max_steph_16_17-1.15, y=steph_2016_17.iloc[max_steph_16_17].SUM_3P + 1, s=(str(steph_2016_17.iloc[max_steph_16_17]['3P']) + " vs " + str(steph_2016_17.iloc[max_steph_16_17]['Opp'])), color="#BF7D14", fontsize=12)

plt.text(x=0, y =-7, s="instagram.com/bballytics", fontsize=14)
plt.text(x=16, y=-7, s="Data: basketbal-reference.com", fontsize=14)

# plt.savefig("harden_three_point_race.jpg", bbox_inches='tight', facecolor='white', edgecolor='white')
```
