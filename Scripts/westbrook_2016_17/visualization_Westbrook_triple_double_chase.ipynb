```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
westbrook_stats = pd.read_csv('../../Data/westbrook_2016_17/westbrook_2016_17.csv')
```

```python
westbrook_stats
```

```python
def truncate_float(number):
    return "%.2f" % number
```

```python
pts, rbs, ast = westbrook_stats.PTS, westbrook_stats.TRB, westbrook_stats.AST
pts_average, rbs_average, ast_average = [], [], []
for i in range(len(pts)):
    if i == 0:
        pts_average.append(truncate_float(pts[i]))
        rbs_average.append(truncate_float(rbs[i]))
        ast_average.append(truncate_float(ast[i]))
    elif i == 80:
        pts_average.append(truncate_float(pts[:(i-1)].mean()))
        rbs_average.append(truncate_float(rbs[:(i-1)].mean()))
        ast_average.append(truncate_float(ast[:(i-1)].mean()))
    else:
        pts_average.append(truncate_float(pts[:i].mean()))
        rbs_average.append(truncate_float(rbs[:i].mean()))
        ast_average.append(truncate_float(ast[:i].mean()))

westbrook_stats.loc[:, "AvgPTS"] = pts_average
westbrook_stats.loc[:, "AvgTRB"] = rbs_average
westbrook_stats.loc[:, "AvgAST"] = ast_average
```

```python
plt.figure(figsize=(12, 12))

plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgPTS.astype(float), label="Points per game", color="#c5423d")
plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgTRB.astype(float), label="Rebounds per game", color="#4341c2")
plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgAST.astype(float), label="Assist per game", color="#40c23b")
plt.xlabel("Game in season")
plt.ylabel("PTS/TRB/AST per game")

plt.title("Russell Westbrook's averages during his 2016-17 NBA season")

plt.legend()
plt.xlim(0, 84)
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(14)
fig.set_figheight(14)

ax.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgTRB.astype(float), label="Rebounds per game", color="#4341c2")
ax.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgAST.astype(float), label="Assist per game", color="#40c23b")
ax.set_xlabel("Game in season")
ax.set_ylabel("TRB/AST per game")

ax.set_title("Russell Westbrook averages during his 2016-17 NBA season")

ax_pts = ax.twinx()

ax_pts.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgPTS.astype(float), label="Points per game", color="#c5423d")
ax_pts.set_ylabel("PTS per game", color='#c5423d')
ax_pts.tick_params("y", colors="#c5423d")

ax.legend(bbox_to_anchor=(1.222,0.98))
ax_pts.legend(bbox_to_anchor=(1.20,1))
ax.set_xlim(0, 84)
```
